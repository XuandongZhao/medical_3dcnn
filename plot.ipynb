```python
# coding: utf-8
import matplotlib.pyplot as plt
import sys

f = open('./1002_rec_0', 'r')
s_train = []
s_test = []
for line in f:
    if 'Train of epoch' in line:
        s_train.append(line)
    if 'Test of epoch' in line:
        s_test.append(line)

x = min(len(s_train), len(s_test))
train_loss = []
test_loss = []
train_acc = []
test_acc = []
train_sen = []
test_sen = []
train_spe = []
test_spe = []
for i in range(x):
    train_l = float(s_train[i].split(' ')[5])
    test_l = float(s_test[i].split(' ')[5])
    train_a = float(s_train[i].split(' ')[7])
    test_a = float(s_test[i].split(' ')[7])
    train_se = float(s_train[i].split(' ')[9])
    test_se = float(s_test[i].split(' ')[9])
    train_sp = float(s_train[i].split(' ')[11])
    test_sp = float(s_test[i].split(' ')[11])

    train_loss.append(train_l)
    test_loss.append(test_l)
    train_acc.append(train_a)
    test_acc.append(test_a)
    train_sen.append(train_se)
    test_sen.append(test_se)
    train_spe.append(train_sp)
    test_spe.append(test_sp)

f, ((ax1, ax2), (ax3, ax4), (ax5, ax6), (ax7, ax8)) = plt.subplots(4, 2, sharey=False)

f.set_size_inches(12, 15, forward=True)
# ax1.set_ylim([0, 1])
ax1.set_title('Train')
ax1.set_ylabel('loss')
ax1.plot(train_loss, color='b', linewidth=0.5)

ax2.set_title('Test')
ax2.plot(test_loss, color='r', linewidth=0.5)

ax3.set_ylabel('Accuracy')
ax3.set_xlabel('epoch')
ax3.plot(train_acc, color='b', linewidth=0.5)

ax4.set_xlabel('epoch')
ax4.plot(test_acc, color='r', linewidth=0.5)

ax5.set_ylabel('Sensitivity')
ax5.set_xlabel('epoch')
ax5.plot(train_sen, color='b', linewidth=0.5)

ax6.set_xlabel('epoch')
ax6.plot(test_sen, color='r', linewidth=0.5)

ax7.set_ylabel('Specificity')
ax7.set_xlabel('epoch')
ax7.plot(train_spe, color='b', linewidth=0.5)

ax8.set_xlabel('epoch')
ax8.plot(test_spe, color='r', linewidth=0.5)
max_acc = max(test_acc)
max_idx = test_acc.index(max_acc)
res_str = 'Max test accuracy: %.4f index %d sensitivity %.3f specificity %.3f' % (
    max_acc, max_idx, test_sen[max_idx], test_spe[max_idx])
print(res_str)
f.suptitle(res_str)
# f.text(1, 2, res_str)
plt.show()
```
